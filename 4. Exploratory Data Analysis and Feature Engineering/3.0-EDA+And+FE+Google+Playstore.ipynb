#### The above error shows that there is some invalid literal of form "3.0M" which we cannot directly convert to int. Let us check and fix the issue.

```python

## to get number of numeric values
df['Reviews'].str.isnumeric().sum()
```

```python
## create a dataframe with complement of numeric values i.e non numeric values
df[~df['Reviews'].str.isnumeric()]
```

```python

df_copy=df.copy() # creates a deep copy, modifications made to df_copy will not affect the df and vice versa
```

```python

df_copy = df_copy.drop(df_copy.index[10472]) ## drops the rows at given index
```

```python
## Convert Review Datatype to int
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'].isnull().sum()
```

```python
## convert the M units into K units
df_copy['Size'] = df_copy['Size'].str.replace('M', '000')
df_copy['Size'] = df_copy['Size'].str.replace('k','')
df_copy['Size'] = df_copy['Size'].replace('Varies with device', np.nan)
df_copy['Size']=df_copy['Size'].astype(float)  # convert the string integers to float, NaN values are already float type
```

```python
df_copy['Size']
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python

## let us clean the the two columns 'Installs' and 'Price'
chars_to_remove = ['+', ',', '$']
cols_to_clean = ['Installs', 'Price']

for char in chars_to_remove:   ## loops through the list chars_to_remove
    for col in cols_to_clean:  ## loops through the list cols_to_clean
        df_copy[col] = df_copy[col].str.replace(char, '')   ## replace the char by empty string


## or we can use the code below to do the same
# pattern = f"[{''.join(chars_to_remove)}]"
# for col in cols_to_clean:
#     df_copy[col] = df_copy[col].str.replace(pattern, '', regex=True)
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype('int')
df_copy['Price']=df_copy['Price'].astype('float')
```

```python
df_copy.info()
```

```python
## Handling Last update feature
df_copy['Last Updated'].unique()
```

```python
## convert the last updated date into proper format and extract the year, day, month,
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Year'] = df_copy['Last Updated'].dt.year
df_copy['Month'] = df_copy['Last Updated'].dt.month
df_copy['Day'] = df_copy['Last Updated'].dt.day
```

```python

## checking for numeric features, dtype ('O') represents object which is typically used for strings in pandas
numeric_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O'] 

## checking for categorical features
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

# print columns
print('We have {} numerical features: {}'.format(len(numeric_features), numeric_features))
print('\n We have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
## Plot the univariate analysis of numerical features
plt.figure(figsize=(15, 15))  # Create a figure of size 15x15 inches
plt.suptitle("Univariate Analysis of Numerical Features", fontsize=25, fontweight="bold", alpha=0.8, y=1)  # Title of the figure

# Loop through the numerical features and create a KDE plot for each feature
for i in range(len(numeric_features)):
    plt.subplot(5, 3, i+1)  # Specify the subplot index (5 rows, 3 columns)
    sns.kdeplot(x=df_copy[numeric_features[i]], shade=True, color="r")  # Kernel density estimate plot with shading (smoothed version of histplot)
    plt.xlabel(numeric_features[i])  # Label the x-axis with the feature name
    plt.tight_layout()  # Adjust layout to prevent overlap

plt.show()
```

## Which Category has largest number of installations??

```python

df_max_installs = df_copy.groupby(["Category"])["Installs"].sum().sort_values(ascending = False).reset_index()
df_max_installs["Installs"] = df_max_installs["Installs"] / 1000000000   ## bracket-style access
## df_max_installs.Installs = df_max_installs.Installs / 1000000000      ## attribute-style access
df2 = df_max_installs.head(10)   ## pick the top 10 category

plt.figure(figsize=(15, 15))
sns.set_context("talk")   ## makes plot more suitable for presentations(it increases the size of the labels, titles, and lines)
sns.set_style("darkgrid")  ## sets the visual style of the plot

bp = sns.barplot(x="Installs", y = "Category", data = df2, palette="coolwarm")
bp.set_xlabel("No. of installations in billions")
bp.set_ylabel("")
bp.set_title("Most popular categories by number of installations")
plt.show()
```

## Insights
1. Out of all the categories "GAME" has the most number of Installations.
2. With almost 14 Billion Installations GAME is the most popular Category in Google App store

## What are the Top 5 most installed Apps in Each popular Categories ??

```python

df_top_apps = df_copy.groupby(["Category", "App"])["Installs"].sum().sort_values(ascending=False).reset_index()
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")     ## adjust the scale of the plot elements (such as labels, titles, and lines) 
sns.set_style("darkgrid")

plt.figure(figsize=(40, 30))

## looping through the categories and creating bar plots
for i,app in enumerate(apps):
    df2 = df_top_apps[df_top_apps.Category == app]   
    df3 = df2.head(5)  ## getting the top 5 apps
    plt.subplot(4, 2, i+1)
    sns.barplot(data=df3, x="Installs", y="App",palette="viridis")
    plt.xlabel("Installation in Millions")
    plt.ylabel("")
    plt.title(app)

plt.tight_layout()
plt.subplots_adjust(hspace= .3)  ##Adds some extra space (hspace=.3) between the rows of subplots to prevent any overlap between titles and the plots.
plt.show()
```

## Insights
- Most popular game is Subway Surfers.
- Most popular communication app is Skype and others.
- Most popular productivity app is Google Drive.
- Most popular social app is facebook, google, and instagram.

## How many apps are there on Google Play Store which get 5 ratings??

```python

rating = df_copy.groupby(["Category", "Installs", "App"])['Rating'].sum().sort_values(ascending=False).reset_index()

top_rated_apps = rating[rating["Rating"] == 5.0]
print("Number of 5 rated apps is ", top_rated_apps.shape[0])
top_rated_apps.head(1)
```

## Result
- There are 271 five rated apps on Google Play store
- Top most is 'CS & IT interview Questions' from 'Family' Category
